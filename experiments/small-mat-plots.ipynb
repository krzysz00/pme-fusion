# Plotting and analysis for small matrix experiment results

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib
```

```python
DATA = "benchmarks_chol_trsm_small_mats_more_points_1"
```

```python
df_raw = pd.read_csv(DATA, sep='\t', index_col=[0, 1], header=None)
```

```python
df_raw
```

```python
avgs = df_raw.mean(axis=1)
```

```python
stddevs = df_raw.std(axis=1)
```

```python
stddevs
```

```python
medians = df_raw.median(axis=1).rename_axis(["N", "Alg."])
```

```python
medians_sep = medians.unstack(1)
```

```python
medians_norm = medians_sep.apply((lambda c: c / (medians_sep.index ** 2)), axis=0)
```

```python
medians_sep.plot()
```

```python
ax = medians_norm.plot(title="Performance")
ax.set_ylabel("Median time(secs) / N^2")
ax
```

```python
percent_change = (medians_sep["Unfused"] - medians_sep["Fused"])/medians_sep["Unfused"]
percent_change
```

```python
ax = percent_change.plot(title="Percent increase in median average execution time for fused algorithm", ylim=(0.0, 0.10))
ax.set_ylabel("% increase")
ax
```
